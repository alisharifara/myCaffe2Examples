```python
import numpy as np

import skimage
import skimage.io as io
import skimage.transform

%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimage
```

```python
def display_image(img, title=None):
    plt.figure(figsize=(5,5))
    plt.imshow(img)
    plt.axis('on')
    title = title + ' \n shape: ' + str(img.shape)
    
    plt.title(title)
    plt.show()
```

```python
# read the image
img = skimage.io.imread('./images/cat.jpg')

# convert it into float32
img = skimage.img_as_float32(img).astype(np.float32)
```

```python
# color image cuz it has 3 channels (RGB)
# caffe2 2 expects to received image in the BGR format (wee need to flip the channel)
img.shape
```

```python
display_image(img, 'original image - RGB')
```

```python
# caffe2 and Opencv expects to received image in the BGR format (wee need to flip the channel)
img_bgr = img[:,:, (2, 1, 0)]
display_image(img_bgr, 'original image - BGR')
```

```python
# image rotation is important when dealing with images from mobile devices
img_rotated = skimage.io.imread('./images/rotated_cat.jpg')
img_rotated = skimage.img_as_float32(img_rotated).astype(np.float32)
```
